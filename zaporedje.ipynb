```python
import matplotlib.pyplot as plt
import numpy as np


old=[]
fancy=[]
    


xdinate=[]

erthrad=6.378e+6
g=9.81
massearth=6e+24
massofdude=80
gravitic=6.6e-11

maxheit=300000000
ptsongraph=100
djump=maxheit/ptsongraph



def whategral(jump):
    topleft=jump*np.sqrt((erthrad+jump)/erthrad)
    topright=np.sqrt(erthrad*jump)*np.arcsinh(np.sqrt(jump/erthrad))
    bottom=np.sqrt(2*gravitic*jump*massearth)*np.sqrt(erthrad+jump)/np.sqrt(erthrad**2*(erthrad+jump))
    
    rsult=(topleft+topright)/(bottom+0.0000000000000000000000000000000000000000000000000000000001)#ne da se mi z limitami jebat
   # print(topleft," * ", topright ," / " ,bottom," = ", rsult)
    return rsult


for hop in range(ptsongraph):
    old.append(np.sqrt(2*djump*hop/g))
    fancy.append(whategral(hop*djump))
    xdinate.append(hop*djump/1000)

plt.grid()

plt.ylabel("padanja v s")
plt.xlabel("razdalja v km")
plt.plot(xdinate,old,label = "če je fg konstanten")
plt.plot(xdinate,fancy, label ="če fg pada")
plt.legend()
print("spreminjanje padalnega časa odvisno od kje vržeš")
    
```
